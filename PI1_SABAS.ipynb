```python
import ast
```

```python
def extract_name(value):
    if isinstance(value, dict):  # Si ya es un diccionario, extraer el nombre
        return value.get('name', None)
    elif isinstance(value, str) and value.startswith("{"):  # Si es un string tipo diccionario
        try:
            parsed_value = ast.literal_eval(value)  # Convertir el string en diccionario
            return parsed_value.get('name', None)  # Extraer 'name' si existe
        except (ValueError, SyntaxError):  # Capturar errores en la conversión
            return None
    return None  # Si es NaN, float, None u otro tipo, devolver None
```

```python
movies['collection_name'] = movies['belongs_to_collection'].apply(extract_name)
```

```python
def extract_id(value):
    if isinstance(value, dict):  # Si ya es un diccionario, extraer el nombre
        return value.get('id', None)
    elif isinstance(value, str) and value.startswith("{"):  # Si es un string tipo diccionario
        try:
            parsed_value = ast.literal_eval(value)  # Convertir el string en diccionario
            return parsed_value.get('id', None)  # Extraer 'name' si existe
        except (ValueError, SyntaxError):  # Capturar errores en la conversión
            return None
    return None
```

```python
movies['id_collection'] = movies['belongs_to_collection'].apply(extract_id)
movies['id_collection'] = movies['id_collection'].astype('Int64') 
```

```python

def get_principal_actor(cast):
    try:
        cast_list = ast.literal_eval(cast)  
        for actor in cast_list:
            if actor.get('order') == 0:
                return actor.get('name')  
        return None  
    except (ValueError, SyntaxError):
        return None  
```

```python
credits['principal_actor'] = credits['cast'].apply(get_principal_actor)
```

```python
credits
```

```python
def get_director(crew):
    try:
        crew_list = ast.literal_eval(crew)  
        for person in crew_list:
            if person.get('job') == 'Director':
                return person.get('name')  
        return None  
    except (ValueError, SyntaxError):
        return None  
```

```python
credits['director'] = credits['crew'].apply(get_director)
```
